```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
w, h = 100,100;
Adjacency_Matrix = [[random.randint(0,2) for x in range(w)] for y in range(h)] 
```

```python
A1 = array(Adjacency_Matrix)
A1
```

```python
def hits(A):
    n= len(A)
    Au= dot(transpose(A),A)
    Hu = dot(A,transpose(A))
    a = ones(n); h = ones(n)
    for j in range(5):
        a = dot(a,Au)
        a= a/sum(a)
        h = dot(h,Hu)
        h = h/ sum(h)
    
    label = websites[0:n] 
    
    index = np.arange(len(label))
    plt.bar(index, a)
    plt.xlabel('Genre', fontsize=5)
    plt.ylabel('No of Movies', fontsize=5)
    plt.xticks(index, label, fontsize=5, rotation=30)
    plt.title('Authority Score')
    plt.show()
    
    index = np.arange(len(label))
    plt.bar(index, h)
    plt.xlabel('Genre', fontsize=5)
    plt.ylabel('No of Movies', fontsize=5)
    plt.xticks(index, label, fontsize=5, rotation=30)
    plt.title('Hub Scores ')
    plt.show()
```

```python
hits(sample)
```

```python
hits(A1)
```
